```python
import numpy as np
import scipy as sp
import pandas as pd
import random
import matplotlib.pyplot as plt
from collections import Counter
```

```python
import numpy as np
# Set the random seed to a specific value (e.g., 42)

def generate_coin_samples(num_trials, bias: float):
    seed_value = 42
    np.random.seed(seed_value)
    """
    Generate coin flip samples for 'n' trials, with a specified bias if provided.

    Parameters:
    num_trials (int): The number of trials.
    bias (float): The probability of getting a biased coin (0.0 for always fair, 1.0 for always biased).

    Returns:
    list: A list of length 'n' containing the results of each coin flip (either 'H' for heads or 'T' for tails).
    """
    if not (bias >= 0.0 and bias <= 1.0):
        raise ValueError("Bias must be a float value between 0.0 and 1.0")

    # Generate random numbers using numpy
    random_numbers = np.random.random(size = num_trials)

    # Determine the coin flip results based on bias
    coin_samples = ['H' if r < bias else 'T' for r in random_numbers]

    return coin_samples, num_trials

coin_samples, num_trials = generate_coin_samples(1000, 0.4)
print(coin_samples)

```

## Experiment 1 -  Tossing biased coin 1000 times, considering bias of 0.4

```python
coin_samples, num_trials = generate_coin_samples(1000, 0.4)
print(coin_samples)
outcome_counts = Counter(coin_samples)
heads_count = outcome_counts['H']
tails_count = outcome_counts['T']
print(f"Number Heads: {heads_count} and Tails: {tails_count} in {num_trials}")
```

## Experiment 1 -  Plot

```python
# Create a bar plot with different colors for 'Heads' and 'Tails'
labels = ['Heads', 'Tails']
counts = [heads_count, tails_count]
colors = ['r', 'orange']

plt.bar(labels, counts, color=colors)
plt.xlabel("Outcome")
plt.ylabel("Frequency")
plt.title("Coin Flip Distribution", fontweight = 'bold')

plt.show()
```

## Bernoulli Trials experiment

A "Bernoulli trial" is a random experiment or process with two possible outcomes: success and failure. These outcomes are typically denoted as "1" for success and "0" for failure. Each Bernoulli trial is independent of previous and future trials, and the probability of success (denoted as "p") and the probability of failure (denoted as "q," where q = 1 - p) remain constant from trial to trial.

```python
def binomial_exp(num_trials, bias:float):
    coin_samples = np.random.choice(['H', 'T'], size = (num_trials, num_trials), p = [bias, 1 - bias])
    num_heads = np.sum(coin_samples == 'H', axis =1)
    plt.hist(num_heads, bins = range(num_trials+2), align='left', rwidth=0.8, density=True, alpha = 0.7)
    plt.xticks(range(num_trials + 1))
    plt.xlabel('Number of Heads')
    plt.ylabel('Probability')
    plt.title(f'Binomial Distribution: {num_trials} Coin Flips (Bias={bias})', fontweight = 'bold')
    plt.show()
```

## Experiment - 1

```python
binomial_exp(10, 0.5)
## number of trails 10 and bias is 0.5 that is fair coin
```
